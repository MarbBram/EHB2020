# Reformat data

Based on the data gathered from this IoT solution, you can see that you have data for 2 different areas, area1 and area2. To easily compare and analyze this data, you should bring each device into its own column.

To load the data, you specify your own column names, since you did not save these with the data. After loading the data, you need to convert the timestamp. The timestamp is in milliseconds, which we need to tell to_datetime by specifying unit="ms".

You pivot the data and resample the data to 1-minute intervals since for this dataset, a more detailed analysis would not make sense.

### Task 1: Convert the unix-timestamp column ts to datetime using unit ms.

```python
import pandas as pd
df = pd.read_csv("date-time.csv")
```

```python
# Replace the timestamp with the parsed timestamp
df['ts'] = pd.to_datetime(df["ts"], unit="ms")
print(df.head())
```

### Task 2: Pivot the DataFrame. The resulting DataFrame df2 should have ts as index, each unique device used as one column header, and val as values.

```python
# Replace the timestamp with the parsed timestamp
df['ts'] = pd.to_datetime(df["ts"], unit="ms")
print(df.head())

# Pivot the dataframe
df2 = pd.pivot_table(df, columns="device", values="val", index="ts")
print(df2.head())
```

### Task 3: Resample the DataFrame to 1min interval using max() as aggregation function and drop empty rows.

```python
# Replace the timestamp with the parsed timestamp
df['ts'] = pd.to_datetime(df["ts"], unit="ms")
print(df.head())

# Pivot the DataFrame
df2 = pd.pivot_table(df, columns="device", values="val", index="ts")
print(df2.head())

# Resample DataFrame to 1min
df3 = df2.resample("1min").max().dropna()
print(df3.head())

```

## Analyzing Energy counter data
Simply plotting energy counter data will not show anything other than 2 straight lines.

The energy counter gives you a running total, but what you really care about his how much energy was consumed between each measurement.

You need to calculate the difference between 2 consecutive values to see how much energy has been consumed in each interval. Creating the percentage-change from the difference allows for easier comparison and allows new insights.

pandas has already been imported as pd, and matplotlib.pyplot as plt. The DataFrame is available as df (so you continue independent from the previous task)

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("energy1.csv", parse_dates=True, index_col=0)
df.info()
```

### Task 1: Create a new DataFrame called df_diff that contains the difference between values in consecutive rows for all columns.

```python
# Get difference between values
df_diff = df.diff()

# Plot the DataFrame
df_diff.plot()
plt.show()
```

### Task 2: Downsample (the original DataFrame) df into 30 minute bins, using max() and store the result in df_res to remove noise.

```python
# Resample df to 30 minutes
df_res = df.resample('30min').max()

# Get difference between values
df_diff = df_res.diff()

# Plot the DataFrame
df_diff.plot()
plt.show()
```

### Task 3: Calculate the percentage changes between consecutive rows in df_diff and store the result in df_pct.

```python
# Resample df to 30 minutes
df_res = df.resample('30min').max()

# Get difference between values
df_diff = df_res.diff()

# Get the percent changed
df_pct = df_diff.pct_change()

# Plot the DataFrame
df_pct.plot()
plt.show()
```
